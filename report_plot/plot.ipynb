```python
import matplotlib.pyplot as plt

n_estimators=[100,200,500,800,1200]
evaluation_distance=[4.577,4.495,4.442,4.267,4.207]
x_label="n_estimators"
y_label="Evaluation distance between ground truth and predications"
title="Parameters Tuning: n_estimators"
```

```python
fig,ax=plt.subplots()
ax.plot(n_estimators,evaluation_distance)

ax.set(xlabel=x_label, ylabel=y_label,
       title=title)
ax.grid()
plt.show()
```

```python
import numpy as np

fig, ax = plt.subplots()
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange', 'tab:green']

ax.bar(n_estimators, evaluation_distance, label=n_estimators, color=bar_colors, width=80)

ax.set_xlabel(x_label)
plt.xticks(np.arange(min(n_estimators), max(n_estimators) + 1, 100))
ax.set_ylabel(y_label)
ax.set_title(title)
plt.show()
```

```python
n_estimators_as_strings = [str(value) for value in n_estimators]
fig, ax = plt.subplots()

bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange', 'tab:green']

ax.bar(n_estimators_as_strings, evaluation_distance, label=n_estimators_as_strings, color=bar_colors)

ax.set_ylabel(y_label)
ax.set_title(title)

plt.show()
```
